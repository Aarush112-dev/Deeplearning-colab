```python
import numpy as np
```

```python
text = "I relish a good sunset. The dramatic dimming horizon when the skies have been blue. That burst of intense colour that occurs when there are enough clouds to provide a canvas. Or the slow menacing darkness when a stormy afternoon turns into a stormy night. TheyвЂ™re all beautiful in their own way, and they all remind us of the rhythm of the day.Months, weeks, hours, minutes and seconds - these are all humans inventions. But days are more natural. You can see the sunrise and sunset and know that another day has passed. On good days you can watch the sunset and enjoy what is, before it isnвЂ™t. On the not-so-good days you can watch the sunset and know that this day is nearly over, and tomorrow is a fresh start."
```

```python
chars = sorted(list(set(text)))
print(chars)
```

```python
chars_dict = {ch:i for i, ch in enumerate(chars)}
index_dict = {i:ch for i, ch in enumerate(chars)}
```

```python
text_indicies = [chars_dict[ch] for ch in text]
print(text_indicies)
```

```python
sequence_len = 4
x = []
y = []

for i in range(len(text_indicies)-sequence_len):
  x.append(text_indicies[i:i+sequence_len])
  y.append(text_indicies[i+sequence_len])

x_array = np.array(x)
y_array = np.array(y)
```

```python
import tensorflow as tf
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, SimpleRNN
```

```python
neural_model = Sequential([
    SimpleRNN(128, input_shape=(x_array.shape[1],len(chars)),activation="relu"),
    Dense(len(chars),activation="softmax")
])
```

```python
neural_model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
x_onehot = np.zeros((x_array.shape[0],x_array.shape[1],len(chars)))
for i, sequence in enumerate(x_array):
  for p,chars_dict in enumerate(sequence):
    x_onehot[i,p,chars_dict] = 1
```

```python
y_onehot = np.expand_dims(y,axis=-1)
```

```python
neural_model.fit(x_onehot,y_onehot,epochs=20,verbose=1, batch_size=32)
```

```python
chars_dict = {ch:i for i, ch in enumerate(chars)}
index_dict = {i:ch for i, ch in enumerate(chars)}
```

```python
def predict(model,seed,char_dict,i_dict,sequence_len):
  if len(seed) < sequence_len:
    seed = seed.zfill(sequence_len)
  elif len(seed) > sequence_len:
    seed = seed[-sequence_len:]
  print(char_dict)

  seed_indicies = [char_dict[ch] for ch in seed]
  x_seed = np.zeros((1,sequence_len,len(char_dict)))
  for t,index in enumerate(seed_indicies):
    x_seed[0,t,index] = 1
  prediction = model.predict(x_seed)
  predicted_index = np.argmax(prediction,axis=-1)[0]
  print(predicted_index)
  predicted_char = i_dict[predicted_index]
  return predicted_char
```

```python
predict(neural_model,seed="aar",char_dict=chars_dict,i_dict=index_dict,sequence_len=sequence_len)
```
